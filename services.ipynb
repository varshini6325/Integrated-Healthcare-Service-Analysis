# FUTURE PREDICTION OF UPCOMING 12 MONTHS

```python
# Convert the 'month_year' column to datetime format
df['month_year'] = pd.to_datetime(df['month_year'])

# Pivot the data to get a time series format with services as columns
pivoted_data = df.pivot_table(index='month_year', columns='SERVICE', values='count', aggfunc='sum')

# Identify the top 3 services based on average counts
top_services = pivoted_data.mean().nlargest(3).index

# Function to fit the ARIMA model and predict future values
def forecast_service_arima(data, periods=12):
    model = ARIMA(data, order=(1, 1, 1))
    fitted_model = model.fit()
    forecast = fitted_model.forecast(steps=periods)
    return forecast

# Create an empty dictionary to store forecasts
forecasts_arima = {}

# Number of periods to forecast (12 months)
periods = 12

# Generate forecasts for the top 3 services
for service in top_services:
    service_data = pivoted_data[service].dropna()
    forecast = forecast_service_arima(service_data, periods)
    forecasts_arima[service] = forecast

# Plot the forecast for the top 3 services combined in a single plot
plt.figure(figsize=(14, 7))

for service in top_services:
    forecast = forecasts_arima[service]
    historical_data = pivoted_data[service]
    
    # Plot historical data
    plt.plot(historical_data, label=f'Historical Data ({service})')
    
    # Annotate historical data points
    for x, y in zip(historical_data.index, historical_data):
        plt.text(x, y, f'{y:.0f}', ha='right', va='bottom')
    
    # Plot forecast data for the next 12 months only
    last_date = historical_data.index[-1]
    forecast_dates = pd.date_range(start=last_date, periods=periods, freq='M')
    plt.plot(forecast_dates, forecast, label=f'Forecast ({service})', linestyle='--')
    
    # Annotate forecast data points
    for x, y in zip(forecast_dates, forecast):
        plt.text(x, y, f'{y:.0f}', ha='right', va='bottom')

plt.title('Forecast for Top 3 Services (Next 12 Months)')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

INFERENCE:

CBC (Complete Blood Count)

Historical data: Increased from 180 to 393 in 2023-07, then dropped to 222, and finally increased steadily.

Forecast: Will remain around 284 throughout the forecast period.

RBS

Historical data: Increased from 181 to 366 in 2023-07, then dropped to 205, and finally increased gradually.

Forecast: Will increase from 240 to 260 in 2025-07.

Creatinine

Historical data: Gradually increased from 160 to 220.

Forecast: Will experience a slight decrease in 2024-07, then increase steadily until 2025-07.

the graph suggests that all three services are expected to continue growing in the future, but at varying rates. CBC is expected to experience the most significant growth, followed by RBS. Creatinine is expected to grow at a more gradual pace.

# FUTURE PREDICTION OF UPCOMING 12 MONTHS

```python
# Convert 'DATE_OF_BILLING' to datetime format
data['DATE_OF_BILLING'] = pd.to_datetime(data['DATE_OF_BILLING'])

# Identify the top 3 services
top_services = data['SERVICE'].value_counts().head(3).index

# Filter data for the top 3 services
top_services_data = data[data['SERVICE'].isin(top_services)]

# Extract month and year from the date
top_services_data['YearMonth'] = top_services_data['DATE_OF_BILLING'].dt.to_period('M')

# Group by 'YearMonth' and 'SERVICE' to get monthly counts
monthly_top_service_counts = top_services_data.groupby(['YearMonth', 'SERVICE']).size().unstack(fill_value=0)

# Function to fit ARIMA model and forecast future counts
def forecast_service(service_name, data, periods=12):
    model = ARIMA(data[service_name], order=(1, 1, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=periods)
    return forecast

# Forecast for each of the top 3 services for next 12 months
forecasts_12_months = {}
for service in top_services:
    forecasts_12_months[service] = forecast_service(service, monthly_top_service_counts, periods=12)

# Plot the forecasts for 12 months
plt.figure(figsize=(14, 7))
for service in top_services:
    # Plot observed data
    plt.plot(monthly_top_service_counts.index.to_timestamp(), monthly_top_service_counts[service], marker='o', label=f'Observed {service}')
    for i in range(len(monthly_top_service_counts)):
        plt.text(monthly_top_service_counts.index[i].start_time, monthly_top_service_counts[service][i], monthly_top_service_counts[service][i], 
                 ha='right', va='bottom', fontsize=9, color='blue')

    # Plot forecasted data for 12 months
    forecast_index = forecasts_12_months[service].index.to_timestamp()
    forecast_values = forecasts_12_months[service].values
    plt.plot(forecast_index, forecast_values, marker='x', linestyle='--', label=f'12-Month Forecast {service}')
    for i in range(len(forecast_values)):
        plt.text(forecast_index[i], forecast_values[i], int(forecast_values[i]), 
                 ha='right', va='bottom', fontsize=9, color='red')

plt.legend()
plt.title('Top 3 Services 12-Month Forecast')
plt.xlabel('Date')
plt.ylabel('Service Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```
